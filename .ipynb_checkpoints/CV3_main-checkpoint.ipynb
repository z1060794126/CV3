## <font color = black>3.2.2分段仿射扭曲</font>

```python
x,y = np.array(np.random.standard_normal((2,100)))
tri = Delaunay(np.c_[x,y]).simplices

plt.figure()
for t in tri:
    t_ext = [t[0],t[1],t[2],t[0]]
    plt.plot(x[t_ext],y[t_ext],'r')

plt.plot(x,y,'*')
plt.show()
```

```python
#打开图像，并将其扭曲
fromim = np.array(Image.open('sunset_tree.jpg'))
x,y = np.meshgrid(range(5),range(6))
#print(x)
#print(y)
x = (fromim.shape[1]/4)*x.flatten()
y = (fromim.shape[0]/5)*y.flatten()

#三角剖分
tri = warp.triangulate_points(x,y)

#打开图像和目标点
im = np.array(Image.open('turningtorso1.jpg'))
tp = np.loadtxt('turningtorso1_points.txt',dtype=int)    #destination points

#将点转换成齐次坐标
fp = np.vstack((y,x,np.ones((1,len(tp)))))
tp = np.vstack((tp[:,1],tp[:,0],np.ones((1,len(tp)))))
#print(tp,'\n\n')
tp = tp.astype(int)
#print('int\n',tp,'\n\n')
#扭曲三角形

im = warp.pw_affine(fromim,im,fp,tp,tri)
```

```python
def obtain_points(i,j):
    """获得图像i和j的匹配点,j=i+1"""
    li,di = sift.read_features_from_file(featname[i])
    lj,dj = sift.read_features_from_file(featname[j])
    
    matches_ji = matches[i]
    
    fi=[[]for t in range(3)]
    fj=[[]for t in range (3)]
    
    for n,m in enumerate(matches_ji):
        m = int(m)
        if m > 0:
            #plt.plot([locs1[i][0],locs2[m][0]+cols1],[locs1[i][1],locs2[m][1]],'c')
            #plt.plot(locs1[i][0],locs1[i][1],'ob')
            #plt.plot(locs2[m][0]+cols1,locs2[m][1],'ob')
            tempi = [li[m][0],li[m][1],1]
            fi[0].append(tempi[1])
            fi[1].append(tempi[0])
            fi[2].append(tempi[2])
            tempj = [lj[n][0],lj[n][1],1]
            fj[0].append(tempj[1])
            fj[1].append(tempj[0])
            fj[2].append(tempj[2])
    
    return np.array(fi),np.array(fj)
    #return fi,fj
```
